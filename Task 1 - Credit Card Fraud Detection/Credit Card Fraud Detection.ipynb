# Checking for Imbalanced data

```python
df['class'].value_counts()
```

# Storing Target Varible in y and others in x

```python
y = df['class']
y
```

```python
x = df.drop('class', axis=1)
x
```

# Oversampling (to remove Imbalance of data)

```python
x.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE # Uses k mean nearest neighbours to produce distinct yet similar records
```

```python
new_x,new_y = SMOTE().fit_resample(x,y)
```

```python
new_y.shape
```

```python
new_y.value_counts()
```

# Splitting dataset into Testing and Training ones

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y = train_test_split(new_x,new_y,test_size=0.25,random_state=50)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier #Importing Decision Tree Classifier
dt = DecisionTreeClassifier()
dt.fit(train_x,train_y)
```

```python
y_pred = dt.predict(test_x)  # Predicitng fraud on Testing dataset
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score,accuracy_score  # To check model is good or not
```

```python
accuracy_score(test_y,y_pred)
```

```python
precision_score(test_y,y_pred)
```

```python
recall_score(test_y,y_pred)
```

```python
f1_score(test_y,y_pred)
```

# Saving the Decision Tree Classifier Model

We will use Decision Tree Classifier on the entire dataset to train our model.

```python
dt_final = DecisionTreeClassifier()
dt_final.fit(new_x.values,new_y) #Using .values to not fit model with column names for ease to use
```

```python
import joblib
```

```python
joblib.dump(dt_final,'fraud_detector')
```

```python
model = joblib.load('fraud_detector')
```
