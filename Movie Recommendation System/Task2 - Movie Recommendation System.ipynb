# Content Based System


Only keeping necessary columns required for recommendation

```python
movies = df.loc[:,['original_title','cast','genres','director']]
movies
```

```python
movies.info()
```

Replacing '|' by a space for easier prcessing

```python
movies['cast'] = movies['cast'].apply(lambda x: x.replace("|"," "))
movies['genres'] = movies['genres'].apply(lambda x: x.replace("|"," "))
```

```python
movies
```

```python
movies['combined'] = movies['genres'] + " " + movies['cast'] + " " + movies['director']
movies.head()
# Creating a new column for recommendation purposes
```

```python
movies['combined'][0]
```

```python
movies['original_title'] = movies['original_title'].str.lower()
movies.head()
# For making the process of recommendation easier
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

#Importing the necessary modules from sklearn library.
```

```python
cv = CountVectorizer()
count_vectorizer_matrix = cv.fit_transform(movies['combined'])
count_vectorizer_matrix

# Making an instance of Count Vectorizer and applying fit_transform on combined column of movies_tmdb dataframe which will return the frequency of feature names for each movie.
```

```python
count_vectorizer_matrix.toarray()[1]
```

```python
similarity_mat=cosine_similarity(count_vectorizer_matrix)
```

```python
similarity_mat
```

```python
movies[73:84]
```

```python
movie_index = movies.loc[movies['original_title']=='hotel transylvania 2'].index[0] #Taking a sample movie for trial
```

```python
movie_list = list(enumerate(similarity_mat[movie_index])) #Making a list of score of similar movies to the sample movie
```

```python
movie_list = sorted(movie_list , key = lambda x:x[1] ,reverse=True) #Sorting the movies in descending order, as higher the similarity score more similar is the movie w.r.t sample movie.

```

```python
movie_list = movie_list[1:11] #Dropping the first movie in the list, as it is the same movie(sample movie)
```

```python
similarmovies = []
for i in range(len(movie_list)):
    a = movie_list[i][0]
    similarmovies.append(movies['original_title'][a])
```

```python
similarmovies  #We get the similar movies to the sample movie selcted.
```
